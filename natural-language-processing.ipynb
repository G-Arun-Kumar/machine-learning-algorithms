# Classical vs Deep Learning models

Now, we will look with some examples of the above venn diagram covering all parts. How models evloved from those days.

**Some examples below**

1. **If/Else Rules(Chatbot):** It's a chat bot under NLP, and in this type a huge list of questions and answers were recorded. Kind of a frequently asked questions. And when users ask for certain questions, the algorithm will look for certain words in the prerecorded questions and answers and if found it will give the user an answer. But, such mechanical approach to answering questions or chatting with people anything realistic where the questions and answers are there but the users want something tailored to them, something specific or something else but that doesn't fit into that list of questions & answers. This becomes a mess. But that's how we used to create chat bots.

2. **Audio-frequency components analysis(speech recognition):** This is another example of NLP which can be applied to speech recognition. There are certain algorithms which uses non-deep learning technique for speech recognition. In essence, we will look at sound waves of somebody talking or prerecorded audio or the real-time audio of a person and then we try to identify what wave forms exist in there. In a simple way, there will be a fourier transformation and there we got the frequency. So, one of these sound waves are travelling in the transformation, and then we compare the audio speech with the sound waves. Each sound waves has got some frequencey and if the wave falls in that frequency, then we find out this type of word or that type of word. associated with those waves. So, the key point is **we are not doing any neural computations. we aren't creating any neural networks.** We are just creating mathematical calculations around the frequencies that we can observe compared to the mathematical calculations we have in our library of pre-analyzed fequencies. When the frequencies are matched, we find out what words are spoken. That's how we recognize the speech.

3. **Bag of words model (Classification)** - It's the very popular approach for text analysis or natural language processing. For example, we have some texts with a pass/fail grade(1-pass and 0-fail grade). Now, this bag of words model will put every texts(great job-1, amazing work-1, poor effort-0) texts are put into a bag and then remembers it. How often the word 'great' come up 1 and how often with 0. This is the high level explanation. But the crux is, it will look at the words and classify or it will try to associates these words with positive or negative results. In our case, the amazing would be associated with positive. And we don't often see 'amazing' when the sentiment is negative. When next time something comes up, it will analyze these words, it will understand these words are associated with 0s or 1s. then it will predict and able to clasify the new comment or text and will say the grade as 1 or 0.

4. **CNN for text recognition (Classification)** - This falls under the category of deep natural language processing. CNN is a neural network that is used for image recognition. For example, a self driving car to detect obstacle or people on roads. Let's see how CNN is used in text recognition. The **words** are transformed in a matrix and that's done through an operation called **embedding of words**. Once in a matrix, the same principle as convolution is applied for image, is applied for text as well. The convolutional operations are done for these images. Then they are **pooled, max pooled, min pooled or some pooled** and then **flattened**. Then we have the prediction.

5. **Seq2Seq (many applications)** - This is the very important model of all the above. Very interesting and cutting edge model. There are many more applications that can be built using this model.
