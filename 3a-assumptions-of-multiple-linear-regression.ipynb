# Assumptions of Linear Regression

Before getting into multiple linear regression, we need look into this topic-assumptions of linear regression. Refer **below picture** for better understanding.

This generally covers the topic of when should we use linear regression model that best suits the data set to be fit. 
The linear regression model will not always be best fit or really serving it's purpose for our data set.Sometimes it may be misleading. So we must be careful before building linear regression model. And that's exactly where we need to consider these assumptions. Refer 'Anscombe's quartet' diagram from document for better clarity.

There are 5 assumptions in total + 1 check. They are,

    1. Linearity
    
    2. Homoesceadasticity (equal variance)
    
    3. Multivariate normality (normality of error distribution)
    
    4. Independence of observations (includes no autocorrelation)
    
    5. Lack of multicollinearity (Predictors are not correlated with each other)
    
    6. Outlier(check)
    
   

1. Linearity: We make sure there is a linear relationship between our dependent variable and each independent variable.

2. Homoesceadasticity : This simply means equal variance. Here,the variance is dependent on independent variable which makes the chart to be in a cone shape.Either increasing cone or decreasing cone. If this occurs, then we **must not apply linear regression** for our dataset.

3. Multivariable normality (normality of error distribution): Here, the data points will not be touching the regression line. If looked along the line of linear regression, we must see a normal distribution of the data point. If not, then we musn't apply linear regression.

4. Independence of observations (includes no autocorrelation): Here, we shouldn't see a kind of pattern in our data. Which means, the rows are not independent. Some rows are affecting other rows and the other are affecting some other rows. A classic example would be **stock market.** In which case, the price of previous stock is affected by future and that is affected by next future stock etc. In this case, we wouldn't apply linear regression.

5. Lack of multicollinearity (Predictors are not correlated with each other): Basically the independent variables or predictors not to be correlated with each other. If **not correlated, we can build the linear regression**. If **correlated and build the linear regression**, then the coefficient estimates we get in model will be unreliable.

6. Outlier(check) : This is an extra check before building linear regression model. The chart clearly shows that the outlier is significantly affecting the linear regression line. So we need to consider whether to remove the outlier before building linear regression or to have the outlier included in the linear regression model.

# Buliding a model

There are 5 different ways to build a model. They are,

1. All-in
2. Backward elimination
3. Forward selection
4. Bidirectional elimination
5. Score comparison

Sometime people refer the **steps 2,3 and 4 as stepwise regression**. Also, some people refer just the **step 4 as stepwise regression**

1. **All-in** - Here, we throw in all your variables. Meaning, we use all the variables to build the model. Conversly, we just discussed that we should not use all the variables that we have. Yet, we use all variables because of 3 reasons.
    a. We have prior knowledge of it-meaning, we have knowledge of the domain and the data we use or these exact variables        are the one are true predictors.
    b. You have to - This means, you have to make use of the variables by the framework followed in organisation. Kind of          compulsion. You actually a different idea but you have to use all of those variables.
    c. Preparing for backward elimination - When we are planning to prepare the backward elimination, we use all of those          variables
    
2. **Backward elimination** - This is the important step and actually the step by step method is followed here. Here, we have 5 steps to fit the model. Below are those.
    **Step 1**: Select the **significance level**. By default, we select it as 0.05(5%).
    
    **Step 2**: Fit in the model with all possible predictors. This means, using the above point 'All in'. We will consider all             the variables (predictors) to work for the model building
    
    **Step 3**: Consider a predictor with highest p-value. If the **p-value > SL** (significance level), then proceed to step 4.               Otherwise go to FIN. Meaning, if the p-value is < SL (p-value smaller than significance level), we are done. we             finished building the model.
    
    **Step 4**: Since, p-value > SL, we need to remove that predictor which has highest p-value.
    
    **Step 5**: Fit the model without that predictor (variable with highest p-value). **And then go back to step 3**
    


3. **Forward selection** - This is quite opposite to backward elimination. But very tedious than that. Below are the steps.

    **Step 1**: Select the significance level. By default it's 0.05
    
    **Step 2**: Fit all simple linear regression model. It means, take a dependent variable and create a regression model                   for every single independent variable. Now, out of all those models we created, we select the one which has                 the lowest p-value for the independent variable.
    
    **Step 3**: We keep this variable that we have chosen and fit all other possible models with an extra predictor added                   to the one we already have. Now we need to construct all possible linear regression with 2 variables where                 one already we have selected.
    
    **Step 4**: Out of all of these possible 2 variable regressions, we consider the one where the new variable has lowest                 p-value. If **p-value < SL, move back to step 3**. Means, now we have the regression with 2 variables and                   now we will add a third varaible.we will try all possible variables that we have left as our third                         variable. And then out of all those models with 3 variables, we will goto step 4 and select again the one                   with the lowest p-value for that third variable we added. We keep doing this and growing a variable at a                   time and then we **stop only when the variable added greater than SL**. Now the tricker part is - we keep                   not the current model but the previous one. because we have just added a variable which is insignificant.                   so just move back.
    
    
   
4. **Bidirectional elimination** - This involves both backward elimination & forward selection. Below are the steps.

     **Step 1** - Select the significance level to enter and stay in the model. Eg. SLENTER = 0.05 and SLSTAY = 0.05
   
     **Step 2** - Perform the next step of forward selection (new variable must have P < SLENTER to enter)
   
     **Step 3** - Perform all steps of backward elimination (old variables must have P < SLSTAY to stay) and then move back                   to step 2. This is an iterative process and you keep doing this until at some point you cannot add new                     variables.
   
     **Step 4** - No variables can enter or no old variables can exit. As soon as we get there, then we proceed to the                       finish because our model is ready. we cannot add anything and we cannot take anything out. This means we                   have created the model.
           
5. **All possible models** - This is probably the most thorough approach and perhaps the resource consuming approach.

      **Step 1** - We select a criterion of goodness of fit (eg. Akaike criterion) can be the R-squared.
    
      **Step 2** - Construct all possible regression models. Suppose we have 'n' variables, then we have (2 power n)-1                        total combinations of these variables and that's exactly how many models there can possibly be.
             
      **Step 3** - Select the one of these models with the best criterion that we are looking at. Now the model is ready.
    
      
**Remember, backward elimination is the fastest way to build the model of all of the 5 types. This is what we are going to see in our chapter**.
