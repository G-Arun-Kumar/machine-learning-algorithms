# Regression model selection

While we had come to the end of the regression session, it's important to know, of these many models in regression,
which one to use? Which model should I apply for my dataset?

**Answer-** In short, apply the model selection technique in all the models, compare the R^2 value and choose the best one. We have learnt about the R^2 value in previous tutorial(another notebook) formula & explanation is available. Just reiterrating here, better our model, smaller will be the SSreg(regression sum of squares) and larger will be the R^2 value. It means, the R^2 value **closer to 1 are with greater performance model**.

We have a function called **r2_score** from the module **metrics** from sklearn library. This **r2_score** function helps to identify the performace of the model and find the R^2 value. By comparing this R^2 value of every model, we can choose the best model.
