# Hierarchical Clustering Intuition

If we have the datapoints or observation points on the scatter plot and apply **Hierarchical clustering**, the output or the result will be very similar to **K-Means Clustering**. But the process is completely different from each other.

There are 2 types of Hierarchical clustering. They are,

**1. Agglomerative clustering**

**2. Divisive clustering**

**Agglomerative** will be the bottom up approach. We will start at the very bottom and then building our way up.

**Divisive** will be quite opposite. It's starting at the top and devising the clusters into multiple.

**We will only look in depth about agglomerative approach**.

Let's look at the step by step approach of agglomerative hierarchical clustering.

**Step1** Make each data point a single cluster --> This forms **N** clusters. So, we have N datapoints and we will look each datapoint as a cluster. which means N cluster.

**Step2** Take two closest data points and make them one cluster --> This forms **N-1 clusters**. So here, we see the datapoints one with other and those that are close with one another will be made as a cluster. They will be combined into one cluster. **Remember all other datapoints that are far away are still considered as a single cluster as per step1**.

**Step3** Take two closest clusters and make them one cluster --> This forms **N-2 clsuters**. Here, we will see two closer clusters (each will be a cluster with 2 datapoints and not a single datapoint cluster) and make them one cluster.

**Step4** Repeat Step 3 until there is only one cluster.

If you see above steps, gradually all single datapoints (individual clusters) are grouped and made as clusters. Gradually the clusters count is reduced. Finally, leading to one large cluster.

The **important point** in above step to note is **closest cluster**. How do we define the closeness? How do we measure the closeness and come to a conclusion that these are closer clusters. Well, the **answer** is using **eucledian distance**. Hope the eucledian distance between two points are already known.

**sqrt((X2-X1)^2+(Y2-Y1)^2)**

The calculation of **distance between two clusters** are very crucial point because it determines our business problem or have a huge impact in business or data science. the calculation could be of many options. They are below.

1.closest distance of datapoint between two clusters

2.farthest distance of datapoint between two clusters

3.Average distance of all datapoints in one cluster to another cluster

4.Distance between centroids.

So, the choice of choosing the calculation is left to individual scenario. But the calculation makes a vital role.



So, the **sole purpose** of doing above steps is the way that the hierarchical clustering works is that **it maintains a memory of how we went through this process** and that process is **stored in dendogram**

# How do dendograms work

**Dendograms** helps us to identify the actual clustering. It means, how many clusters we would have at the end. This is what we need to identify so that we have our datapoints being clustered which paves the way for further analysis. This is what we have done in K-Means clustering. We will identify clusters.

So, dendograms help us to identify the actual no.of clusters.

Below diagram has charts. On the left is the chart which contains 6 points. On the right we have got another chart and we will use it for **dendogram creation**. The right chart is first added with the points bigger that are from left chart. The right chart has **euclidean distance** on y-axis and points on **x-axis**.

Now, on the left chart, we will follow the steps of **agglomerative hierarchical clustering** by forming clusters of two closest datapoints.  

Since **Dendogram** is the memory of hierarchical clustering, we need to signify the left side chart cluster in right side dendogram.
So we will connect the two points (P2 & P3) in right side dendogram which are clusters of left chart. We will do this by **drawing the horizontal line** on these two points. And we will draw at the height of exactly the distance of P2 & P3 in left chart.
Eg. If the distance b/w P2 & P3 is 5, then we will draw an horizontal line at the height of 5 from bottom(zero) in dendogram.

Simply, the euclidean distance of two points is the height of dendpgram points. This also represents the **computed dissimilarities between the two points or two clusters.**

Then we move on to next two closest clusters and connect them. And then again we draw the horizontal line of those two points in dendogram. The height of dendogram of those two points are measured by the distance of those two points in left chart.

Now, from below digram, the points P1 and (P2&P3 cluster) are now closer. So, we will unite them. And the distance between P1 & P3 are measured and drawn a vertical line with same height. The horizontal line is drawn above P1 and (P2&P3 cluster) to connect both. Refer below green color dendogram points of P1,P2 & P3.

Same way, P4 is connected to (P5&P6 cluster) and drwan the dendogram. Refer red color dendogram of P4, P5 & P6.

Finally combining all the cluster into one single cluster and drawn the dendogram. If you see, the construction of dendogram is **done gradually from bottom up** hence the name **bottom up approach of agglomerative hierchical clustering**. Also, the memory is mainitained, if you could see the diagram, the euclidean distance of two points are always maintained.

# Get optimal no.of clusters using dendogram and Hierchical Clustering

Let's use the same example and chart and diagram for understanding.

How to use the dendogram to understand how to get the best Hierchical Clustering.

**Answer** We can set a **threshold** at horizontal levels of dendogram. This can be said be height threshold or distance threshold or dissimiliarities threshold because the vertical axis measures the median distance between two points or clusters.

For instance, **we set a threshold at 1.7 in euclidean distance (y-axis), then we will have only 2 clusters**. It's nothing but we don't need any cluster beyond that threshold. So, the horizontal line that cuts or intersects the dendogram is the count of clusters. Since this intersects or cuts two lines of dendogram, then we have **2 clusters** for our algorithm. This is the value we apply in algorithm to identify the dependent variable that will give us future clusters. In other words, this is the value where the datapoints are segmented or grouped of classified into patterns with clusters for further analysis.

----->Diagram with 4 dendograms.

Simply **the no.of clusters are identified by the horizontal threshold that is set in dendogram. The count of vertical line that cuts the horizontal threshold lines is the count of clussters.**

When a threshold value is set, it means we don't want clusters that have any points within them that have dissimilarity greater than this threshold.

So that's how the dendograms work by setting the threshold, we can find the optimal clusters that suits best our project.
