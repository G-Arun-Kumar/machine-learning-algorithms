# Kernel SVM Intuition

So, we know how SVMs work and find the prediction boundry or the prediction curve that seperates or splits the class of data or two categories of data so that if a new observations are found, it will exactly identify the class and classify it accordingly.

Now, what if the situation comes **where the prediction curve or a boundary cannot be drawn**.
What if the prediciton border line cannot be drawn or **the data didn't allow us to find the prediction curve**.

Sometimes, data (both type of classes-red & green) are mingled with each other. Or encircled with each other like a **concentric circles**. We cannot define the prediction curve at this situation. This situation (data mingled with each other) is due t0 **non-linear type of data**. The data is not linear. Hence, we cannot draw a prediction boundry.



In SVM, since the data is linear or the SVM assumes the data is linear, we can define the prediction curve. Then, how to define the prediction curve if the data is non-linear?

That is where **Kernel SVMs comes in**. First, we will explore a method called **higher dimensional space** where we will learn how we can take this dataset and add an extra dimension in the space we are dealing with **and make our data linearly seperable.** We will see this with couple of examples. Then we will discuss **kernel trick** which allow us to do this in a **more computationally efficient way without having higher dimensional space** Later on we will discuss another types of kernel as well. 

# A Higher dimensional space

Now we will take our non-linear seperable dataset, map it to a higher dimension and get a linearly seperably data, invoke the SVM algorithm, build a decision boundary for our dataset and project back to the original dimension.

First we will see the 1-dimensional dataset. Here we have got 9 non-lineardatapoints mixed with red and green in a one dimension X1. Now, how to change the non-linear dataset into linear dataset by increasing the dimensionality of space.

**Answer -** We will do this by **creating a mapping function**. Let's say we use **f=x-5 as a mapping function**, the entire elements in the dimension will move towards left with negative signs (-2,-3,-4) and few with positive left (2,3,4) and then the next step would be **squared of that. means, f=(x-5)^2**. Now we **have a squared function** going through the chart and everything will be projected onto the function. 

Now if we try to draw predictive boundry or a linear seperator (**In 2D, it's a straight line**), it forms the linearly seperable data. Now we would project everything back to our original space.

In 3D it's a hyperplane. The linear seperator is a hyperplane.

There maay be a different mapping function as well apart from f=x-5.

Now, for a 2-dimensional space - we will apply any other mapping function to find the linearly seperable dataset.

But there's a **problem with this approach (mapping to a higher dimension) algorithm can be highly compute-intensive**. It has a lot of computation, lots of processing power and therefore this **approach is not the best because the datasets are mapped into higher dimenstion, performing all calculation and coming back to lower dimension** cause a lot of delays and processing backlogs.

So, we will take another approach called as **kernel SVM** and this will perform very similar resulst but without having going to higher dimensions.

Let's now look into the kernel trick without even taking or projecting our dataset into higher dimensional space and doing complex computations etc..

Let's understand the kernel function from the picture first and later relate it with the formula.

We see an x-y coordinate in 2D space in the picture. Now, in the center where x-axis is 0 and y-axis is 0 is where the **landmark is located.** The vertical calibrations in the x plane is the height of landmark and let's see below how the landmark is growing bigger and bigger and reach the top of the hill by relating with the formula.

Let's take a point from our dataset which is placed far away from the **landmark(always placed in the center)**. When we substitute that point in the above formula, **we get a value very close to zero. How?** Because,when we substitute x(datapoint), landmark and formulate, since the datapoint is far away from the landmark, the difference of landmark and point (numerator) is large and squaring will even be more large. Then dividing it by the sigma, will also be large. After taking the negative power to the exponential, final value will still be large. So it will be somewhere close to **zero**.

If we take another datapoint which is closer to landmark, try to substitute in formula, final result will be very very small. **Something like 0.0001483** And so, exponential power to that 0.0001483 is like **exponential to the power of zero will be one**. By this, the kernel function grow bigger and bigger and reach the top of the hill.

From this we infer that, if the datapoint is far away from the landmark, the value will be **zero** And if the datapoint is close to landmark, the value will be **one**.

**Sigma role-** The role of sigma is to correctly define the circumference of the landmark. We will take the circuference of the landmark by correctly adjusting the sigma **and assign value as 1 for all datapoints inside the circle** and **for datapoints outside the circle will be assigned as 0**. 

In this way, we correctly define the value of sigma, draw the circle, assign values and **split the datapoints as green and red**. this is how the decision boundaries are defined using kernel SVM.

This is the **kernel trick** we follow to differentiate the data points and draw the prediction boundaries or **achieve the linearly seperable datapoints without mapping to higher dimensional space**.

Eventhough for complex datapoints, we can achieve the linearly seperable, by adding two kernel functions together.
